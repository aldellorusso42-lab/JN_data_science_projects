```python
import xml.etree.ElementTree as ET
import pandas as pd

xml_string = """
<robot>
    <test name = "Test Login">
        <status status = "PASS" elapsed="1.2"/>
    </test>
    <test name="Test Logout">
        <status status="FAIL" elapsed="0.8"/>
    </test>
    <test name="Test Dashboard">
        <status status="PASS" elapsed="2.4"/>
    </test>
</robot>
"""
tree = ET.fromstring(xml_string)
print(tree)

for test in tree.findall("test"):
    name = test.get("name")
    status = test.find("status").get("status")
    elapsed = test.find("status").get("elapsed")
    print(name, status, elapsed)
```

```python
xml_string = """
<results>
    <module name="Navigation">
        <case name="Load Map">
            <outcome result="PASS" duration="2.1"/>
        </case>
        <case name="Search Location">
            <outcome result="FAIL" duration="1.4"/>
        </case>
    </module>
    <module name="Settings">
        <case name="Update Profile">
            <outcome result="PASS" duration="0.7"/>
        </case>
        <case name="Change Password">
            <outcome result="FAIL" duration="0.5"/>
        </case>
    </module>
</results>
"""

tree = ET.fromstring(xml_string)

rows = []

for x in tree.findall("module"):
    mod_name = x.get("name")
    for y in x.findall("case"):
        case_name = y.get("name")
        result = y.find("outcome").get("result")
        rows.append({"Module Name": mod_name, "Case Name":case_name, "Result":result})

finish = pd.DataFrame(rows)
print(finish)
```

```python
tree = ET.parse("robo_output.xml")
root = tree.getroot()
from datetime import datetime
sets=[]

for a in root.findall("suite"):
    suite_name = a.get("name")
    for b in a.findall("test"):
        tag = b.find("tags").find("tag").text
        test_name = b.get("name")
        status = b.find("status").get("status")
        starttime = b.find("status").get("starttime")
        endtime = b.find("status").get("endtime")
        for c in b.findall("kw"):
            keyword_name = c.get("name")
            sets.append({"Suite Name":suite_name, "Test Name":test_name, "Status": status,
                     "Start Time": starttime, "End Time": endtime, "Keyword Name": keyword_name, "Tag": tag})

t = pd.DataFrame(sets)
t["Start Time"] = pd.to_datetime(t["Start Time"], format= "%Y%m%d %H:%M:%S.%f")
t["End Time"] = pd.to_datetime(t["End Time"], format= "%Y%m%d %H:%M:%S.%f")
t["Duration"] = (t["End Time"] - t["Start Time"]).dt.total_seconds()

print(t)
            
```

```python
xml_string = """
<robot>
    <suite name="Login Tests">
        <test name="Valid Login">
            <status status="PASS" starttime="20200511 15:16:08.428" endtime="20200511 15:16:08.430"/>
            <tags>
                <tag>sanity</tag>
            </tags>
        </test>
        <test name="Invalid Password">
            <status status="FAIL" starttime="20200511 15:16:08.431" endtime="20200511 15:16:08.433"/>
        </test>
    </suite>
</robot>
"""

tree = ET.fromstring(xml_string)
rows = []
for a in tree.findall("suite"):
    suite_name = a.get("name")
    for b in a.findall("test"):
        test_name = b.get("name")
        status = b.find("status").get("status")
        tags_element = b.find("tags")
        if tags_element is not None:
            tag = tags_element.find("tag").text
        else:
            tag = "No Tag"
        rows.append({"Suite":suite_name, "Test":test_name, "Status":status, "Tag":tag})

df = pd.DataFrame(rows)
print(df)
```
