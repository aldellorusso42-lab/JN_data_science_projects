# 2. Background and Objectives

### Background
#### For radio frequency and digital signal processing (referred to as RF and DSP), many signals can be modeled as sine waves which can be write in equation form as: 

#### $ x(t) = Asin(2ft\pi+ Φ) $
#### $A$ = Amplitude (how tall the wave is)
#### $f$ = frequency in Hz (# of cycles per second)
#### $t$ = time in seconds
#### $Φ$ = the phase in radians (where in its cycle the wave starts)

#### By changing amplitude, frequency, and phase, we can see how each property affects the signal which we will need to understand before moving into more complex topics

# 3. Computational Methods

#### We will use the following tools and methods:

#### Tools: NumPy and Matplotlib
#### Methods: Create a function, generate_sine that returns an array that repersents the equation above. Then change the parameters on this function to see how it varies.


#### Basic python code with explainations:

```python
import numpy as np
import matplotlib.pyplot as plt    #Allows us to use both functions

fs = 10000  # sampling rate in Hz
duration = 0.05  # seconds
t = np.linspace(0, duration, int(fs * duration), endpoint=False) # This creates 500 evenly spaced vaules from 0 to 500 without included the endpoint

# Now t is a list of time stamps where we can evaluate our equation

def generate_sine(amplitude, frequency, phase, t):
    """
    This generates a sine wave: x(t) = A * sin(2*pi*f*t + phase)
    with the parameters:
    
    - Amplitude as a float 
    
    - Frequency as a float in Hz

    - Phase as a float in radians

    - Time as an array in seconds
    """
    return amplitude * np.sin(2 * np.pi * frequency * t + phase)

def plot_signal(t, x, title="", xlabel="Time (s)", ylabel="Amplitude"):
    
    plt.figure(figsize=(8, 3))
    plt.plot(t, x)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
# We will use this to show a visual of what happens when we change each variable
```

# 4. Visualizations and Observations

## 4.1 Varying Amplitude

```python
# Here we want to keep frequency and phase as fixed parameters
f = 100  
phase = 0  

amplitudes = [0.5, 1.0, 2.0]
signals = []

plt.figure(figsize=(8, 4))
for A in amplitudes:
    x = generate_sine(A, f, phase, t)
    signals.append(x)
    plt.plot(t, x, label=f"A = {A}")

plt.title("Sine Waves with Different Amplitudes (f = 100 Hz, phase = 0)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.tight_layout()
plt.show()
```

### Observations:
#### - When amplitude increases, the peaks and valleys of the sine wave become taller, but the spacing between peaks (the period) stays the same.
#### - All three waves complete the same number of cycles in the same time interval
#### - In RF terms, a larger amplitude often corresponds to a stronger or more powerful signal.

## 4.2 Varying frequency

```python
A = 1  
phase = 0  

frequency = [50, 100, 300]

plt.figure(figsize=(8, 4))
for f in frequency:
    x = generate_sine(A, f, phase, t)
    plt.plot(t, x, label=f"f = {f} Hz")

plt.title("Sine Waves with Different frequencies (A = 1, phase = 0)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.tight_layout()
plt.show()
```

### Observations: 
#### - Higher frequency waves complete more cycles in the same amount of time, so the waveform looks more compressed horizontally.
#### - The low frequency wave (50 Hz) has wide oscillations, while the 300 Hz wave oscillates rapidly.
#### - The amplitude of the waves is the same, which shows that frequency and amplitude are independent properties.

# 4.3 Varying Phase

```python
f = 100  
A = 1  

phase = [0,  np.pi/4, np.pi/2, np.pi]

plt.figure(figsize=(8, 4))
for phi in phase:
    x = generate_sine(A, f, phi, t)
    plt.plot(t, x, label=f"phase = {phi:.2f} radians")

plt.title("Sine Waves with Different phase (A = 1, f = 100)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.tight_layout()
plt.show()
```

### Observations: 
#### - Changing the phase shifts the waveform left or right in time without changing its amplitude or frequency.
#### - A phase shift of $\pi$ radians flips the wave. Meaning where one wave has a peak, the other has a valley.
#### - For the same amplitude and frequency, multiple sine waves can look identical except for where they start in their cycle.

# 5. Conclusions

We saw that: Amplitude controls how tall the wave is, without changing how often it oscillates. Frequency controls how many cycles occur per second, compressing or stretching the waveform in time. Phase shifts the waveform left or right without changing its height or its oscillation rate.
