# Combinding Signals

# 1. Introduction

#### For this notebook, we will look at how more complex signals can be created by adding simple sine waves together. Real world radio signals come from mixtures of many frequences not just a single tone like we perviously looked at. By visualizing how two-tone and three-tone combinations behave in time, we’ll start to see how more complex waveforms can come from very simple mathematics. 

# 2. Background and Objectives

## Background

#### As we saw previously a sine wave can be written in the form:
#### $ x(t) = Asin(2ft\pi+ Φ) $
#### So if we have to sine waves and add them together we get:
#### $ x(t) = x_1(t) + x_2(t) = A_1sin(2f_1t\pi+ Φ_1) +  A_2sin(2f_2t\pi+ Φ_2) $
#### This is known as a superposition

## Objective

#### 1. Generate individual sine waves.
#### 2. Plot two-tone and three-tone combinations.
#### 3. Compare component signals to the combined signal.
#### 4. Create multi-panel figures showing each signal and the resulting sum.
#### 5. Build comparison tables describing signal characteristics


# 3. Computational Methods

#### We will use NumPy, Matplotlib, and reusable functions to generate and visulaize the experiment. We will also use simple numpy array addition and multi-panel plots to compare the signals

```python
# Same code set up from phase 1
import numpy as np
import matplotlib.pyplot as plt  

fs = 10000 
duration = 0.05 
t = np.linspace(0, duration, int(fs * duration), endpoint=False) 

def generate_sine(amplitude, frequency, phase, t):
    return amplitude * np.sin(2 * np.pi * frequency * t + phase)

def plot_signal(t, x, title="", xlabel="Time (s)", ylabel="Amplitude"): 
    plt.figure(figsize=(8, 3))
    plt.plot(t, x)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

# 4. Visualizations

#### We are now going to create the plots and show what happens when frequencies are mixed.

### 4.1 Two-tone Signals

```python
# Equal Amplitude but different frequencies
A1, f1, phase1 = 1, 100, 0
A2, f2, phase2 = 1, 300, 0

x1 = generate_sine(A1, f1, phase1, t)
x2 = generate_sine(A2, f2, phase2, t)

x_total = x1 + x2

fig, axes = plt.subplots(3, 1, figsize=(10,6), sharex=True)

axes[0].plot(t,x1)
axes[0].set_title("100 Hz Sine Wave")
axes[0].set_ylabel( "Amplitude")

axes[1].plot(t, x2, color="orange")
axes[1].set_title("300 Hz Sine Wave")
axes[1].set_ylabel("Amplitude")

axes[2].plot(t, x_total, color="green")
axes[2].set_title("100 Hz + 300 Hz")
axes[2].set_xlabel("Time (s)")
axes[2].set_ylabel("Amplitude")

plt.show()
```

### Observations:
#### - Both the 100 Hz and 300 Hz signals are basic on their own but when combind they create a much more complex signals that isn't easy to decypher just by looking at it.
#### - When a peak and valley meet, there difference brings them back to zero.

## 4.2 Two-Tone signals with different Amplitudes and frequencies

```python
A1, f1, phase1 = 0.5, 100, 0
A2, f2, phase2 = 2, 300, 0

x1 = generate_sine(A1, f1, phase1, t)
x2 = generate_sine(A2, f2, phase2, t)

x_total = x1 + x2

fig, axes = plt.subplots(3, 1, figsize=(10,8), sharex=True)

axes[0].plot(t,x1)
axes[0].set_title("100 Hz, Amplitude = 0.5")
axes[0].set_ylabel( "Amplitude")
axes[0].set_ylim(-2,2)

axes[1].plot(t, x2, color="orange")
axes[1].set_title("300 Hz, Amplitude = 2")
axes[1].set_ylabel("Amplitude")

axes[2].plot(t, x_total, color="green")
axes[2].set_title("100 Hz + 300 Hz")
axes[2].set_xlabel("Time (s)")
axes[2].set_ylabel("Amplitude")

plt.show()
```

### Observations: 
#### - Since the 300 Hz signal has a higher frequency and a higher Amplitude, it dominates the output when the two signals are added together.
#### - Every time the 0.5 amplitude signal has a peak or valley it is off set by the other signal

## 4.3 Three-Tone Signal

```python
# Three different signals all with different frequencies and Amplitudes
A1, f1, phase1 = 1, 50, 0
A2, f2, phase2 = 0.5, 175, 0
A3, f3, phase3 = 0.75, 300, 0

x1 = generate_sine(A1, f1, phase1, t)
x2 = generate_sine(A2, f2, phase2, t)
x3 = generate_sine(A3, f3, phase3, t)

x_total = x1 + x2 + x3

fig, axes = plt.subplots(4, 1, figsize=(10,10), sharex=True)

axes[0].plot(t,x1)
axes[0].set_title("50 Hz, Amp = 1")
axes[0].set_ylabel( "Amplitude")

axes[1].plot(t, x2, color="orange")
axes[1].set_title("175 Hz, Amp = 0.5")
axes[1].set_ylabel("Amplitude")

axes[2].plot(t, x3, color="blue")
axes[2].set_title("300 Hz, Amp = 0.75")
axes[2].set_ylabel("Amplitude")

axes[3].plot(t, x_total, color="green")
axes[3].set_title("175 Hz + 300 Hz + 50 Hz")
axes[3].set_xlabel("Time (s)")
axes[3].set_ylabel("Amplitude")

axes[0].set_ylim(-2,2)
axes[1].set_ylim(-2,2)
axes[2].set_ylim(-2,2)

plt.show()
```

### Observations: 
#### - Even though the sum is formed from three simple sine waves, the waveform is very complex looking
#### - There is still a pattern to the waveform but it takes longer to repeat due to it having to calculate three waves at once

# 5. Conclusions

#### We learned that superposition means the total signal is created by the sum of its components.
